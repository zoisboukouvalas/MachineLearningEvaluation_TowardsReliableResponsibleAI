```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate data randomly distributed from a 2nd degree polynomial 
np.random.seed(42)
x = 5*np.random.rand(20)
x = np.linspace(0,5,20)
x = np.sort(x)
x = np.reshape(x,[-1,1])
y = 2*x**3 + 3*x + 25*np.random.normal(loc=0, scale=1, size=[20,1])
```

```python
# Create a figure and scatter plot using the grayscale colormap
plt.figure(figsize=(10, 7), dpi=80)  # Set the background color to black
plt.scatter(x, y, lw=2.5, c='k')

plt.show()
```

```python
# Linear Regression
model_deg1 = LinearRegression()
model_deg1.fit(x,y)
y_pred_1 = model_deg1.predict(np.reshape(x, [-1,1]))

# 3rd Degree Polynomial
model_deg3 = LinearRegression()
p_3 = PolynomialFeatures(degree=3, include_bias=False)
p_3.fit(x.reshape([-1,1]))
model_deg3.fit(p_3.transform(x.reshape([-1,1])),y)
y_pred_3 = model_deg3.predict(p_3.transform(x.reshape([-1,1])))

# 12 degree Polynomial
model_deg20 = LinearRegression()
p_20 = PolynomialFeatures(degree=20, include_bias=False)
p_20.fit(x.reshape([-1,1]))
model_deg20.fit(p_20.transform(x.reshape([-1,1])),y)
y_pred_20 = model_deg20.predict(p_20.transform(np.linspace(0,5,1000).reshape([-1,1])))
```

```python
plt.figure(figsize = (10,7), dpi=80)
plt.scatter(x,y, linewidths=4, color="k")
plt.plot(x,y_pred_1, color="k", label = "Linear", lw=2.5, )
plt.plot(x,y_pred_3, color="k", label = "3rd Degree Polynomial", lw=2.5, ls ="--")
plt.plot(np.linspace(0,5,1000),y_pred_20, color="k", label ="20th Degree Polynomial", lw=2.5, ls =":")
plt.legend()
plt.savefig("./Figures/Figure2-1.pdf",format="pdf")
plt.show()
```
