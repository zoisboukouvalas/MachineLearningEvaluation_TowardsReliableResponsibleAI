```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Generate data randomly distributed from a 2nd degree polynomial 
np.random.seed(42)
x = np.linspace(1,5, 100)
x = np.reshape(x,[-1,1])
y = 2*x**3 + 3*x + 25*np.random.normal(loc=0, scale=1, size=[100,1])
```

```python
plt.figure(figsize = (10,7), dpi=80)
plt.scatter(x,y)
plt.show()
```

```python
# Linear Regression
model_deg1 = LinearRegression()
model_deg1.fit(x,y)
y_pred_1 = model_deg1.predict(np.reshape(x, [-1,1]))

# 3rd Degree Polynomial
model_deg3 = LinearRegression()
p_3 = PolynomialFeatures(degree=3, include_bias=False)
p_3.fit(x.reshape([-1,1]))
model_deg3.fit(p_3.transform(x.reshape([-1,1])),y)
y_pred_3 = model_deg3.predict(p_3.transform(x.reshape([-1,1])))

# 12 degree Polynomial
model_deg20 = LinearRegression()
p_20 = PolynomialFeatures(degree=20, include_bias=False)
p_20.fit(x.reshape([-1,1]))
model_deg20.fit(p_20.transform(x.reshape([-1,1])),y)
y_pred_20 = model_deg20.predict(p_20.transform(np.linspace(1,5,1000).reshape([-1,1])))
```

```python
plt.figure(figsize = (10,7), dpi=80)
plt.scatter(x,y)
plt.plot(x,y_pred_1, color="r")
plt.plot(x,y_pred_3, color="g")
plt.plot(np.linspace(1,5,1000),y_pred_20, color="k")
# plt.legend()
plt.show()
```
