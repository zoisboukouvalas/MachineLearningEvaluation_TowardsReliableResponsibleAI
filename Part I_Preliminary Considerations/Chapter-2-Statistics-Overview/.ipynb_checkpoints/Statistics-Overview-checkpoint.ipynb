## 2.1 Random Variables

### 2.1.2 Example

#### Prepare Data

[Link to Dataset](https://archive.ics.uci.edu/ml/datasets/Labor+Relations)

```python
import pandas as pd
import numpy as np

# Get Data:
def read_in(link, row_start, nrows):
    return pd.read_csv(link, skiprows=row_start, nrows=nrows, on_bad_lines="skip", delim_whitespace=" ",
                       index_col="#", na_values="*")

# Training Set
link = "https://archive.ics.uci.edu/ml/machine-learning-databases/labor-negotiations/labor-negotiations.data"
good1 = read_in(link,row_start=152, nrows=18)
good2 = read_in(link,row_start=173, nrows=18)
good3 = read_in(link,row_start=194, nrows=18)
good = good1.join(good2)
good = good.join(good3)
bad1 = read_in(link,row_start=216, nrows=9)
bad2 = read_in(link,row_start=228, nrows=9)
bad3 = read_in(link,row_start=240, nrows=9)
bad = bad1.join(bad2)
bad = bad.join(bad3)
bad.index = np.arange(19,28)

train = pd.concat([good,bad])
train["status"] = train.index <= 18
train.head()
```

```python
# Test Set
link = "https://archive.ics.uci.edu/ml/machine-learning-databases/labor-negotiations/labor-negotiations.test"
good1 = read_in(link,row_start=157, nrows=19)
good2 = read_in(link,row_start=179, nrows=19)
good3 = read_in(link,row_start=201, nrows=19)
good = good1.join(good2)
good = good.join(good3)

bad1 = read_in(link,row_start=223, nrows=11)
bad2 = read_in(link,row_start=237, nrows=11)
bad3 = read_in(link,row_start=251, nrows=11)
bad = bad1.join(bad2)
bad = bad.join(bad3)
bad.index = np.arange(20,31)

test = pd.concat([good,bad])
test["status"] = test.index <= 19
```

```python
print(train.shape)
```

```python
print(test.shape)
```

#### Combine Train and Test

```python
from sklearn.preprocessing import OneHotEncoder
from numpy import concatenate

combined = pd.concat([train,test]).reset_index().drop("index",axis=1)# Combine train and test
tmp = combined.select_dtypes(include="object") # Convert to one hot
one_hot = pd.get_dummies(tmp)


#X = combined[["dur","wage1"]].drop(41).to_numpy("int32") # NaN in the row 41
tmp = combined.drop("status", axis=1).select_dtypes(include="float64")
```

#### Handle Empty Values

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# These values being missing is indicative of them being absent from the agreements.
needs_zeros = ["dur","wage1", "wage2", "wage3", "stby_pay", "shift_diff"]
# Only 6 are missing.  In the cases where they are missing, it is more likely actually missing than actually zero.
needs_avg = ["hours"]
# Today in Canada there are 10 federal work holidays.  Some of the entries have only 9, so I will impute that value.
needs_min = ["holidays"] 

imp_zeros = SimpleImputer(missing_values=float("nan"), strategy = "constant", fill_value=0)
imp_avg = SimpleImputer(missing_values=float("nan"), strategy="mean")
imp_min = SimpleImputer(missing_values=float("nan"), strategy = "constant", fill_value=9)

imp = ColumnTransformer(transformers=[("zeros", imp_zeros, needs_zeros), ("avg", imp_avg, needs_avg), 
                                     ("min", imp_min, needs_min)])
imp.fit(tmp)
tmp = imp.transform(tmp)
```

#### Combine Numerical and Categorical

```python
from sklearn.preprocessing import scale
tmp = scale(tmp)
X = concatenate((tmp,one_hot),axis=1)
y = combined[["status"]].to_numpy("int32")
```

#### Train/Test Splits and 10-fold CV Setup

```python
from sklearn.model_selection import ShuffleSplit
from numpy.random import seed, choice

# Split is 6 test, 50 train
seed(42)
ind_list = []
for trial in range(10):
    ind = []
    for i in range(10):
        ind.append(choice(55,6,replace=False))
    ind_list.append(ind)
```

#### Running Classifiers and Storing Errors

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import numpy as np

errors_full = np.empty((0,10))
errors_p_full = np.empty((0,10))
for trial in range(10):
    errors = np.zeros((3,10))
    errors_p = np.zeros((3,10))
    for i in range(10):
        test_idx = list(ind_list[trial][i])
        train_idx = list(set(test_idx) ^ set(range(56)))
        X_test = X[test_idx,:]
        y_test = y[test_idx]
        X_train = X[train_idx,:]
        y_train = y[train_idx]
        DT = DecisionTreeClassifier()
        LR = LogisticRegression()
        SV = SVC()
        predictions = DT.fit(X_train, y_train.ravel()).predict(X_test)
        errors[0,i] = len(y_test) - sum(sum(predictions == y_test.T))
        errors_p[0,i] = errors[0,i]/len(y_test)
        predictions = LR.fit(X_train, y_train.ravel()).predict(X_test)
        errors[1,i] = len(y_test) - sum(sum(predictions == y_test.T))
        errors_p[1,i] = errors[1,i]/len(y_test)
        predictions = SV.fit(X_train, y_train.ravel()).predict(X_test)
        errors[2,i] = len(y_test) - sum(sum(predictions == y_test.T))
        errors_p[2,i] = errors[2,i]/len(y_test)
        
    errors_full = np.concatenate((errors_full, errors))
    errors_p_full = np.concatenate((errors_p_full, errors_p))
```

```python
np.savetxt("ClassifierError.csv", errors_full, delimiter = ",")
```

```python
pd.set_option("display.precision", 0)

Model_names = ["Decision Tree", "Logistic Regression", "SVM"] * 10 
Trial_names = sum([["Trial " + str(1*n),  " ", " "] for n in range(1,11)], [])
fold_names = ["f" + str(num) for num in range(1,11)]
results_df = pd.DataFrame(errors_full, columns=fold_names, index=Trial_names)
results_df.insert(0,"Model",Model_names)
results_df["Total"] = results_df[fold_names].sum(axis=1)
results_df
```

```python
pd.set_option("display.precision", 3)
Model_names = ["Decision Tree", "Logistic Regression", "SVM"] * 10 
fold_names = ["f" + str(num) for num in range(1,11)]
results_df_p = pd.DataFrame(errors_p_full, columns=fold_names, index=Trial_names)
results_df_p.insert(0,"Model",Model_names)
results_df_p
```

### (Chapter 4) t-Test Example Using Above Results

```python
from scipy.stats import ttest_rel

# H0: The error of Logistic Regression and Decision Tree across folds was the same.
# Ha: The error of Logistic Regression is less than the error of Decision Trees across folds.
```

**For One Trial**

```python
dt = results_df_p.iloc[0,1:11].values
svm = results_df_p.iloc[2,1:11].values


ttest_rel(dt, svm, alternative="greater")
```

#### Confidence Intervals

```python
import scipy.stats as st

DT = errors_p_full[0:31:3,:].reshape(100,1)
LR = errors_p_full[1:31:3,:].reshape(100,1)
SVM = errors_p_full[2:31:3,:].reshape(100,1)

CI_dt = list(st.t.interval(alpha=0.95, df=len(DT)-1, loc=np.mean(DT), scale=st.sem(DT)))
CI_lr = list(st.t.interval(alpha=0.95, df=len(DT)-1, loc=np.mean(LR), scale=st.sem(LR)))
CI_svm = list(st.t.interval(alpha=0.95, df=len(DT)-1, loc=np.mean(SVM), scale=st.sem(SVM)))
```

```python
CI_dt
```

```python
import matplotlib.pyplot as plt
plt.hist(DT)
```

```python
import matplotlib.pyplot as plt
def plot_CI(x, CI):
    top = CI[1]
    bottom = CI[0]
    width = 0.1
    left = x - width/2
    right = x + width/2
    plt.plot([x,x], [top, bottom], color="k", lw=3)
    plt.plot([left,right], [top,top], color="k", lw=3)
    plt.plot([left,right], [bottom,bottom], color="k", lw=3)
    plt.plot(x,(top-bottom)/2 + bottom, "ok", lw=3)
```

```python
plt.figure(figsize=(10,10))
plt.xticks([1, 2, 3], ['Decision Tree', 'Logistic Regression', 'SVM'], fontsize=14)
plt.yticks(fontsize=14)
plot_CI(1,CI_dt)
plot_CI(2,CI_lr)
plot_CI(3,CI_svm)
plt.ylabel("Error", fontsize=14)
plt.savefig("./Figures/Figure1-4.pdf",format="pdf")
plt.show()
```

### 2.2.2 The Binomial Distribution

```python
import numpy as np

p=0.5
n=1000
M = [5,10,100, 1000]
array = np.zeros((n,len(M)))

np.random.seed(42)
for m in range(len(M)):
    for i in range(n):
        array[i,m] = np.random.binomial(M[m],p)

# Plotting
plt.figure(figsize=(12, 6), dpi=80)

plt.subplot(1,5,1)
plt.title("m=5")
plt.hist(array[:,0])

plt.subplot(1,5,2)
plt.title("m=10")
plt.hist(array[:,1])

plt.subplot(1,5,3)
plt.title("m=100")
plt.hist(array[:,2],bins=20)

plt.subplot(1,5,4)
plt.title("m=1000")
plt.hist(array[:,3], bins=20)


plt.tight_layout()

plt.savefig("./Figures/Figure1-2.pdf",format="pdf")
plt.show()
```

```python
import numpy as np

P = [0.3,0.8]
n= 1000
m = 100
array = np.zeros((n,len(P)))

np.random.seed(42)
for p in range(len(P)):
    for i in range(n):
        array[i,p] = np.random.binomial(m,P[p])

# Plotting
plt.figure(figsize=(12, 6), dpi=80)

plt.subplot(1,2,1)
plt.hist(array[:,0], edgecolor="black")
plt.title("p=0.3")

plt.subplot(1,2,2)
plt.hist(array[:,1], edgecolor="black")
plt.title("p=0.8")

plt.tight_layout()
plt.savefig("./Figures/Figure1-3.pdf",format="pdf")
```

# 2.4 Hypothesis Testing

```python
from scipy.stats import norm
plt.figure(figsize=(10,10))
n=200
r=6
fs=14
plt.plot(np.linspace(-r,r,n), norm.pdf(np.linspace(-r,r,n), loc=1.5,scale=2),lw=2, color="k")
plt.plot(np.linspace(-r,r,n), norm.pdf(np.linspace(-r,r,n), loc=-1.5,scale=1.5), 'k',lw=2)
plt.plot([0.5,0.5],[0,0.3], 'k')
plt.plot([-r,r],[0,0], 'k',lw=0.7)
plt.fill_between(np.linspace(-r,0.5,200), norm.pdf(np.linspace(-r,0.5,200), loc=1.5,scale=2),alpha=0.25,color="k",
                hatch = "\\")
plt.fill_between(np.linspace(0.5,r,200), norm.pdf(np.linspace(0.5,r,200), loc=-1.5,scale=1.5),alpha=0.755,color="k",
                hatch = "//")
plt.text(-2, 0.27, "H0 Mean", fontsize=fs, color="k")
plt.text(1, 0.21, "H1 Mean", fontsize=fs, color="k")
plt.text(-1, 0.05, "Beta", fontsize=fs)
plt.text(1, 0.01, "Alpha", fontsize=fs,color = "w")
plt.savefig("./Figures/Figure1-5.jpg",format="jpg", dpi=500)
plt.show()
```
